# Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find
i. Sum of all elements
ii. Sum of all elements row-wise
iii. Sum of all elements column-wise *italicized text*

```python
import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
total_sum = np.sum(gfg)
print("Total Sum:", total_sum)
row_sum = np.sum(gfg, axis=1)
print("Row-wise Sum:\n", row_sum)
col_sum = np.sum(gfg, axis=0)
print("Column-wise Sum:\n", col_sum)
```

#Q.2 (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find
i. Sorted array
ii. Indices of sorted array
iii. 4 smallest elements
iv. 5 largest elements *italicized text*

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
print("Sorted Array:\n", sorted_array)
indices = np.argsort(array)
print("Indices of Sorted Array:\n", indices)
smallest_4 = sorted_array[:4]
print("4 Smallest Elements:\n", smallest_4)
largest_5 = sorted_array[-5:]
print("5 Largest Elements:\n", largest_5)
```

#Q2. (b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find
i. Integer elements only
ii. Float elements only

```python
assd = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer_elements = assd[assd == assd.astype(int)]
print("Integer Elements:\n", integer_elements)
float_elements = assd[assd != assd.astype(int)]
print("Float Elements:\n", float_elements)
```
